```python
#Replacing spaces with underscore
df.columns=df.columns.str.replace(" ","_")
df.columns
```

```python
#Let us use median value for missing data in the rating column
rating_median=df["Rating"].median()
print(rating_median)
df["Rating"].fillna(rating_median,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Data Preprocessing
df["Reviews"]=df["Reviews"].astype("int64")
df.info()
```

```python
df["Reviews"].describe().round()
```

```python
#Handling the size column:
df["Size"].unique().size
```

```python
df["Size"].unique()
```

```python
df["Size"]=df["Size"].str.replace("M","",regex=True)
df["Size"]=df["Size"].str.replace("k","",regex=True)
```

```python
df["Size"].unique()
```

```python
sm=df[df["Size"]!="Varies with device"]["Size"].astype(float).median()
```

```python
sm
```

```python
df["Size"]=df["Size"].replace("Varies with device",sm)
```

```python
df["Size"]
```

```python
df["Size"]=pd.to_numeric(df["Size"])
```

```python
# Handling Install column
df["Installs"].unique()
```

```python
df["Installs"]=df["Installs"].apply(lambda x:x.replace("+",""))
df["Installs"]=df["Installs"].apply(lambda x:x.replace(",",""))
df["Installs"]=df["Installs"].apply(lambda x:int(x))
```

```python
df["Installs"].unique()
```

```python
#Handling price column:
df["Price"].unique()
```

```python
df["Price"]=df["Price"].str.replace("$","")
```

```python
df["Price"].unique()
```

```python
df["Price"]=pd.to_numeric(df["Price"])
```

```python
df["Price"].unique()
```

```python
df["Genres"].unique().size
```

```python
df["Genres"].head(10)
```

```python
df["Genres"]=df["Genres"].str.split(";").str[0]     #str[0] is index 0
```

```python
df["Genres"].unique().size
```

```python
df["Genres"].value_counts()
```

```python
df["Genres"].replace("Music & Audio","Music",inplace=True)
```

```python
df["Genres"].value_counts()
```

```python
df["Last_Updated"].head()
```

```python
df["Last_Updated"]=pd.to_datetime(df["Last_Updated"])
```
