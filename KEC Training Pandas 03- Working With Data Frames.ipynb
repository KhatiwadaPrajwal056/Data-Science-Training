# Working with Data Frames

# First import Pandas

```python
import pandas as pd
```

# Creates a dictionary called 'data' that contains information about individuals' names, scores, sports, and gender.

```python
data = {
    "names" : ["Bill", "Tom", "Tim", "John", "Alex", "Vanessa", "Kate"],
    "score" : [90, 80, 85, 75, 95, 60, 65],
    "sport" : ["Wrestling", "Football", "Skiing", "Swimming", "Tennis", "Karate", "Surfing"],
    "sex" : ['M', 'M', 'M', 'M', 'F', 'F', 'F']
}
data
```

# Create a pandas DataFrame object named 'df' using the dictionary 'data' that was defined earlier.

```python
df = pd.DataFrame(data)
df
```

# Retrieve the first few rows of the DataFrame 'df'.

```python
df.head(3)
```

# Retrieve the last few rows of the DataFrame 'df'.

```python
df.tail()
```

# Create a pandas DataFrame named 'df' using the dictionary 'data' and specifies the column names as ["name", "sport", "gender", "score", "age"]. Simply add Age to the existing data frame.

```python
df= pd.DataFrame(data, columns= ["names", "sport", 'gender', "score", "age"])
df
```

# Create a pandas DataFrame named 'df' using the dictionary 'data' and specifies both the column names and the index labels as [one , two ,three, four. etc].

```python
df= pd.DataFrame(data, columns= ["names", "sport", 'gender', "score", "age"],
                index = ["one", "two", "three", "four", "five", "six", "seven"])
df
```

# Retrieve the column labeled "sport" from the DataFrame 'df'.

```python
df['sport']
```

# Retrieve specific columns from the DataFrame 'df' based on the list of column names provided in the variable my_columns.

```python
my_columns = ["names", 'sport']
df[my_columns]
```

# Retrieve a specific row from the DataFrame 'df' based on the provided index label "one".

```python
df.loc[["one"]]
```

# Retrieve multiple rows from the DataFrame 'df' based on the provided index labels "one" and "two".

```python
df.loc[["one", "two"]]
```

# Assign the value 18 to a new column named "age" in the DataFrame 'df'.

```python
df["age"]=18
df 
```

# Assign the values from the list values to the "age" column in the DataFrame 'df'.

```python
values= [18, 19, 20, 18, 17, 17, 18]
df["age"] = values
df
```

# Create a new column named "pass" in the DataFrame 'df' based on the condition df.score>=70. The resulting column should contain boolean values (True or False) indicating whether each corresponding score in the "score" column meets the condition.

```python
df["pass"] = df.score>=70
df
```

# Remove the column named "pass" from the DataFrame 'df'

```python
del df["pass"]
df
```
