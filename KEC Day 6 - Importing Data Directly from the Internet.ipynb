# Import Data directly from Internet and perform EDA

# Import necessary Libraries

```python
import pandas as pd
```

# Fetches the CSV data from the provided URL and stores it in the namesdata variable.
URl  = https://github.com/fivethirtyeight/data/tree/master/unisex-names
    Take (unisex_names_table.csv)

```python
url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/unisex-names/unisex_names_table.csv"
namesdata = pd.read_csv(url)
namesdata
```

# Only fetch the top 5 rows for now

```python
namesdata.head()
```

# Remove The Unamed: 0 column from the dataframe

```python
namesdata = namesdata.drop(columns = ["Unnamed: 0"])
namesdata
```

# Sort the namesdata DataFrame based on the values in the 'gap' column and displays the sorted DataFrame  to see which names have the male and female share almost the same.

Check if your analysis is correct : https://fivethirtyeight.com/features/there-are-922-unisex-names-in-america-is-yours-one-of-them/

```python
namesdata = namesdata.sort_values(by=["gap"])
namesdata
```
