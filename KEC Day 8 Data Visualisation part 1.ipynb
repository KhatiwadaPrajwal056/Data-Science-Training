# Plotting graph with plot() method


# Generate a random series of numbers using NumPy's random number generator. The numbers are drawn from a standard normal distribution.  Plot a line graph using the given series

```python
data = pd.Series(np.random.randn(1000).cumsum())
data
data.plot()
```

# Create a DataFrame named df1 using the pandas library. The DataFrame should have 100 rows and 4 columns. The values in the DataFrame are generated using NumPy's random number generator, drawn from a standard normal distribution. 

```python
df1 = pd.DataFrame(np.random.randn(100,4), columns = list("ABCD"))
df1.head()
```

# Calculate the cumulative sum along each column of the DataFrame

```python
df1 = df1.cumsum()
df1.head()
```

# Plot the line graph using df1 dataframe

```python
df1.plot()
```

# Select the row at index 10 from the DataFrame df1 using the iloc function. Create a bar plot of the values in that row.

```python
df1.iloc[10].plot(kind="bar")
```

```python
df1.iloc[10].plot.bar()
```

# Create a DataFrame named df2 using the pandas library. The DataFrame should have 7 rows and 3 columns. The values in the DataFrame are generated using NumPy's random number generator, drawn from a uniform distribution between 0 and 1.

```python
df2 = pd.DataFrame(np.random.rand(7,3), columns = list("ABC"))
df2
```

# Making Histograms

# Import "Iris" dataset from your machine to your notebook and assign relevant headers to the data.

```python
iris = pd.read_csv("Data/iris.data", header = None)
iris
```

```python
iris.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris.head()
```

# Plot the histogram of iris dataset

```python
iris.plot.hist(alpha = 0.5)
```

# Now try to make a stacked histogram using iris data

```python
iris.plot.hist(alpha = 0.5, stacked= True)
```

# Generate a histogram plot for the DataFrame or Series named "iris". Set the transparency of the bars in the histogram to 1, meaning they should be fully opaque. Stack the bars on top of each other in the histogram plot. Histogram should be divided into 25 bins, or intervals, to represent the range of values in the data

```python
iris.plot.hist(alpha = 1, stacked= True, bins = 25)
```

# Display the chart horizontally

```python
iris.plot.hist(alpha = 1, stacked= True, bins = 25, orientation = "horizontal")
```

# Generate a horizontal histogram plot for the "sepal_width" column of the DataFrame or Series named "iris".

```python
iris["sepal_width"].plot.hist(orientation = "horizontal")
```

# Plot a histogram to show the frequency or count of the differences between consecutive sepal lengths.

```python
iris["sepal_length"].diff().hist()
```

# Generate a histogram plot for each column in the DataFrame "iris".

```python
iris.hist(alpha = 0.5, color="red", bins= 20)
```

# Generate separate histograms for the "petal_length" column of the DataFrame "iris", grouped by the "species" column.

```python
iris.hist("petal_length", by ="class")
```

# Making BoxPlots

# Generates a box plot for each column in the DataFrame "iris".

```python
iris.plot.box()
```

# Define a dictionary named colors with different keys representing different elements of a box plot: 'boxes', 'whiskers', 'medians', and 'caps'. Generate a box plot for each column in the DataFrame "iris" with customized colors assigned in that dictionary

```python
colours = {'boxes': "Red" , 'whiskers':'blue', 'medians': "black", 'caps': "green"}
iris.plot.box(color = colours)
```

# Generate horizontal box plots for each column in the DataFrame "iris".

```python
iris.plot.box(vert = False)
```

# Generate separate box plots for each column in the DataFrame "iris", grouped by the "species" column. It also applies specific styling settings to the plots.

```python
plt.rcParams["figure.figsize"] = [6,6] 
plt.style.use("ggplot")
iris.boxplot(by='class')
```
