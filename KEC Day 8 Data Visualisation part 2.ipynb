# Making Hexagonal bin charts

# Making Pie charts

# First lets calculate the average (mean) value of the "petal_width" column in the DataFrame "iris" for each unique species of iris.

```python
iris_avg = iris["petal_width"].groupby(iris["class"]).mean()
iris_avg
```

# Generate a pie chart using the Series "iris_avg". The values from the Series are used to determine the sizes of the pie slices, and the unique species names from the index are used as labels for each slice.

```python
iris_avg.plot.pie()
```

# Calculate the average (mean) values of the "petal_width" and "petal_length" columns in the DataFrame "iris" for each unique species of iris.

```python
iris_avg1 = iris[["petal_width", "petal_length"]].groupby(iris["class"]).mean()
iris_avg1
```

# Generate multiple pie charts using the DataFrame "iris_avg_2". Each pie chart should represent the distribution of average petal width and petal length for each species of iris.

```python
iris_avg1.plot.pie(subplots = True)
```

# Generate a pie chart using the Series "iris_avg" with customized settings. The values from the Series are used to determine the sizes of the pie slices, and the specified labels, colors, fontsize, and figsize are used for customization.

```python
iris_avg.plot.pie(labels = ["setosa", "versicolor", "virginica"],
                 colors = list("brg"), fontsize= 25, figsize = (10,10))
```

# Generate a pie chart using the Series "iris_avg" with further customizations. The values from the Series are used to determine the sizes of the pie slices, and the specified labels, colors, autopct, fontsize, and figsize are used for customization.

```python
iris_avg.plot.pie(labels = ["setosa", "versicolor", "virginica"],
                 colors = list("brg"), fontsize= 25, figsize = (10,10), autopct = '%.2f')
```
